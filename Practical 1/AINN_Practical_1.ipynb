***AND Gate Truth Table***

```python
xA = [0,1,0,1,0,1,0,1]
xB = [0,0,1,1,0,0,1,1]
xC = [0,0,0,0,1,1,1,1]
T = 2
w = 1
print("AND Gate Truth Table: \n")
print(f"| x1 | x2 | x2 | w | net |O/p|")

for i in range(0, len(xA)):
  net = xA[i]*w + xB[i]*w + xC[i]*w
  op = 0
  if net > T:
    op = 1
  else:
    op = 0
  print(f"| {xA[i]} | {xB[i]} | {xC[i]} | {w} | {net} | {op} |")
```
